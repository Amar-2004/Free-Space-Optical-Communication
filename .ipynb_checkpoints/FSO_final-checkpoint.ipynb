Needed Modules and Constants

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfc
# Parameters
Pt = 100  # Transmitted power in mW
d1 = 0.1  # Distance d1 in meters
d2 = 0.1  # Distance d2 in meters
distances = np.linspace(100, 5000, 50)  # Distances from 100 to 5000 meters
theta = 1  # Beam divergence in mrad
P_noise = 1e-9  # Noise power in mW

# Weather conditions
weather_conditions = {
    'Clear': 0.1,  # Clear weather attenuation in km^-1
    'Rainy': 1.0,  # Rainy weather attenuation in km^-1
    'Foggy': 5.0   # Foggy weather attenuation in km^-1
}

ber_threshold = 1e-3
```

Received Power

```python
def calculate_received_power(Pt, d1, d2, L, theta, beta):
    # Convert L from meters to kilometers
    L_km = L / 1000.0
    
    # Calculate the received power using the provided formula
    Pr = Pt * (d2 / (d1 + (L_km * theta)))**2 * np.exp(-beta * L_km)
    
    return Pr
```

SNR calculations

```python
def calculate_snr(Pr, P_noise):
    # Calculate SNR in linear scale and then convert to dB
    snr_linear = Pr / P_noise
    snr_db = 10 * np.log10(snr_linear)
    return snr_db
```

Bit Error Rate(BER) Calculations

```python
def calculate_ber(snr_db):
    # Convert SNR from dB to linear scale
    snr_linear = 10 ** (snr_db / 10)
    # Calculate BER using the Q-function approximation
    ber = 0.5 * erfc(np.sqrt(snr_linear))
    return ber
```

FSO switching with BER

```python
def fso_rf_switch(Pt, d1, d2, distances, theta, beta, P_noise, ber_threshold=1e-3):
    results = {'distance': [], 'Pr_fso': [], 'snr_fso': [], 'ber_fso': [], 'link_type': []}
    
    for L in distances:
        Pr_fso = calculate_received_power(Pt, d1, d2, L, theta, beta)
        snr_fso = calculate_snr(Pr_fso, P_noise)
        ber_fso = calculate_ber(snr_fso)
        
        if ber_fso > ber_threshold:
            link_type = 'RF'
        else:
            link_type = 'FSO'
        
        results['distance'].append(L)
        results['Pr_fso'].append(Pr_fso)
        results['snr_fso'].append(snr_fso)
        results['ber_fso'].append(ber_fso)
        results['link_type'].append(link_type)
    
    return results
```

Simulation And Plotting

```python
# Parameters
Pt = 100  # Transmitted power in mW
d1 = 0.1  # Distance d1 in meters
d2 = 0.1  # Distance d2 in meters
distances = np.linspace(100, 5000, 50)  # Distances from 100 to 5000 meters
theta = 1  # Beam divergence in mrad
P_noise = 1e-9  # Noise power in mW

# Weather conditions
weather_conditions = {
    'Clear': 0.1,  # Clear weather attenuation in km^-1
    'Rainy': 1.0,  # Rainy weather attenuation in km^-1
    'Foggy': 5.0   # Foggy weather attenuation in km^-1
}

ber_threshold = 1e-3

# Simulation and plotting
fig, ax = plt.subplots(3, 1, figsize=(10, 15), sharex=True)

for i, (weather, beta) in enumerate(weather_conditions.items()):
    results = fso_rf_switch(Pt, d1, d2, distances, theta, beta, P_noise, ber_threshold)
    
    ax[i].plot(results['distance'], results['Pr_fso'], label='Received Power (FSO)', color='blue')
    ax[i].plot(results['distance'], results['snr_fso'], label='SNR (FSO)', color='orange')
    ax[i].plot(results['distance'], results['ber_fso'], label='BER (FSO)', color='green')
    ax[i].fill_between(results['distance'], 0, 1, where=np.array(results['link_type'])=='RF', color='red', alpha=0.3, label='Switch to RF')
    
    ax[i].set_title(f'FSO Link Performance in {weather} Weather')
    ax[i].set_xlabel('Distance (m)')
    ax[i].set_ylabel('Value')
    ax[i].set_yscale('log')
    ax[i].legend()
    ax[i].grid(True)

plt.tight_layout()
plt.show()
```
